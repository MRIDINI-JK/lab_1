```python
import random
import time
import matplotlib.pyplot as plt

def search_in_sequence(sequence, value):
    return value in sequence

def generate_random_sequence(length):
    return random.sample(range(1, length+1), length)

def plot_time_complexity(x, y_start, y_middle, y_end):
    plt.plot(x, y_start, marker='o', label='Start')
    plt.plot(x, y_middle, marker='o', label='Middle')
    plt.plot(x, y_end, marker='o', label='End')
    plt.xlabel('Sequence Length')
    plt.ylabel('Average Time (seconds)')
    plt.title('Time Complexity of Search Positions')
    plt.legend()
    #plt.grid(True)
    plt.show()

sequence_lengths = [5000,10000,15000,20000]  # Vary the sequence lengths as desired
num_trials = 1000

search_positions = ['Start', 'Middle', 'End']
search_values = {position: [] for position in search_positions}

for length in sequence_lengths:
    for position in search_positions:
        values = [random.choice(generate_random_sequence(length)) for _ in range(num_trials)]
        search_values[position].append(values)

average_times = {position: [] for position in search_positions}

for position in search_positions:
    for length, values in zip(sequence_lengths, search_values[position]):
        total_time = 0
        for value in values:
            sequence = generate_random_sequence(length)
            sequence.sort()  # Simulating unordered sequence

            start_time = time.time()
            search_in_sequence(sequence, value)
            end_time = time.time()

            total_time += (end_time - start_time)
        average_time = total_time / num_trials
        average_times[position].append(average_time)

plot_time_complexity(sequence_lengths, average_times['Start'], average_times['Middle'], average_times['End'])
```

```python
import random
import time
import matplotlib.pyplot as plt

# Function to perform linear search
def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

# Generate a list of n random integers
n = 10000
arr = [random.randint(0, n*2) for _ in range(n)]

# Define keys for best, average, and worst cases
best_case_key = arr[0]            # Element at first position
average_case_key = arr[n // 2]    # Element at middle
worst_case_key = -1               # Element not in list

# Measure time for best case
start_time = time.time()
linear_search(arr, best_case_key)
best_time = (time.time() - start_time) * 1000  # ms

# Measure time for average case
start_time = time.time()
linear_search(arr, average_case_key)
average_time = (time.time() - start_time) * 1000  # ms

# Measure time for worst case
start_time = time.time()
linear_search(arr, worst_case_key)
worst_time = (time.time() - start_time) * 1000  # ms

# Print execution times
print("Execution Times (in milliseconds):")
print(f"Best Case:    {best_time:.4f} ms")
print(f"Average Case: {average_time:.4f} ms")
print(f"Worst Case:   {worst_time:.4f} ms")

# Plotting
cases = ['Best Case', 'Average Case', 'Worst Case']
times = [best_time, average_time, worst_time]
colors = ['green', 'orange', 'red']

plt.figure(figsize=(8, 5))
plt.bar(cases, times, color=colors)
plt.ylabel("Time (ms)")
plt.title(f"Linear Search Time Analysis (n = {n})")
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

# Modified linear search that counts comparisons
def linear_search_with_comparisons(arr, key):
    comparisons = 0
    for i in range(len(arr)):
        comparisons += 1
        if arr[i] == key:
            return i, comparisons
    return -1, comparisons

# Different input sizes to test
input_sizes = [100, 500, 1000, 5000, 10000]

# Lists to store comparisons for each case
best_case_comparisons = []
average_case_comparisons = []
worst_case_comparisons = []

# Perform tests for each input size
for n in input_sizes:
    arr = [random.randint(0, n*2) for _ in range(n)]

    # Best case: key is the first element
    _, best_comps = linear_search_with_comparisons(arr, arr[0])

    # Average case: key is the middle element
    _, avg_comps = linear_search_with_comparisons(arr, arr[n // 2])

    # Worst case: key is not in the array
    _, worst_comps = linear_search_with_comparisons(arr, -1)

    best_case_comparisons.append(best_comps)
    average_case_comparisons.append(avg_comps)
    worst_case_comparisons.append(worst_comps)

# Plotting comparisons vs input size
plt.figure(figsize=(10, 6))
plt.plot(input_sizes, best_case_comparisons, marker='o', label='Best Case (O(1))', color='green')
plt.plot(input_sizes, average_case_comparisons, marker='o', label='Average Case (O(n/2))', color='orange')
plt.plot(input_sizes, worst_case_comparisons, marker='o', label='Worst Case (O(n))', color='red')
plt.title('Linear Search: Comparisons vs Input Size')
plt.xlabel('Input Size (n)')
plt.ylabel('Number of Comparisons')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
